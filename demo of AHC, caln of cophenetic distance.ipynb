```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Define the dataset
data = np.array([[2, 4],
                 [3, 6],
                 [5, 8],
                 [9, 7],
                 [12, 2]])

# Perform hierarchical clustering
Z = linkage(data, method='single')  # Using single linkage (shortest distance)

# Generate dendrogram
plt.figure(figsize=(8, 6))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Point Index')
plt.ylabel('Distance')
plt.show()

# Display distance matrix
distance_matrix = np.around(Z[:, 2], decimals=2)
print("Distance Matrix:")
print(distance_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

data = np.array([[2, 4],
                 [3, 6],
                 [5, 8],
                 [9, 7],
                 [12, 2]])

# Perform agglomerative hierarchical clustering
Z = linkage(data, method='single')  # Using single linkage method

# Plot the dendrogram
plt.figure(figsize=(8, 6))
dendrogram(Z)
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.title('Agglomerative Hierarchical Clustering Dendrogram')
plt.show()

# Extract the merging steps
merging_steps = []
for step in Z:
    cluster1 = int(step[0])
    cluster2 = int(step[1])
    merged_cluster = int(step[3])
    merging_steps.append((cluster1, cluster2, merged_cluster))

# Print the merging steps
for i, step in enumerate(merging_steps):
    print(f"Step {i+1}: Merge clusters {step[0]} and {step[1]} into cluster {step[2]}")
```

```python
import numpy as np

# Generate the data
data = np.array([[2, 4],
                 [3, 6],
                 [5, 8],
                 [9, 7],
                 [12, 2]])

# Calculate the pairwise distances
distances = np.zeros((5, 5))

for i in range(5):
    for j in range(5):
        distances[i, j] = round(np.linalg.norm(data[i] - data[j]), 1)

# Display the distance matrix
for i in range(5):
    for j in range(5):
        print(distances[i, j], end="\t")
    print()
```
